```python
pal = sns.color_palette("Paired")
```

```python
cmaps = {
    'alexnet': {
        'transfer': {
            'unit_both': pal[0],
            'mn_both': pal[1],
        },
        'initialize': {
            'unit_both': pal[2],
            'mn_both': pal[3],            
        }
    },
    'CORnet_Z': {
        'transfer': {
            'unit_both': pal[4],
            'mn_both': pal[5],
        },
        'initialize': {
            'unit_both': pal[6],
            'mn_both': pal[7],            
        }
    }
}
```

```python
net_name = 'alexnet'
method = 'transfer'
learning_rate = 1e-03

datasets = [
    '12000samples',
    '12800samples_balanced',
    '3200samples_setsize1',
]
```

```python
df_2_v_5_dict = {}

stim_abbrev = '2_v_5'
for dataset in datasets:
    data_csv_path = CSV_DIR.joinpath(
        f'{net_name}_{stim_abbrev}_{dataset}_split.csv')
    results_gz_path = RESULTS_DIR.joinpath(
        f'{net_name}_{method}_lr_1e-03_no_finetune_{stim_abbrev}_{dataset}/'
        f'searchnets_{net_name}_{method}_lr_1e-03_no_finetune_{stim_abbrev}_{dataset}_'
        'trained_200_epochs_test_results.gz')

    df_2_v_5_dict[dataset] = searchnets.utils.general.results_df(
        data_csv_path, results_gz_path, net_name, method, learning_rate
    )
```

```python
figsize=(12, 8)
fig, ax = plt.subplots(2, 3, sharex=True, sharey=True, figsize=figsize)

FONTSIZE = 24
for axis in ax.ravel():
    for item in ([axis.title, axis.xaxis.label, axis.yaxis.label] +
                 axis.get_xticklabels() + axis.get_yticklabels()):
        item.set_fontsize(FONTSIZE)

# add a big axis, hide frame
fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)

add_legend = False

stim_abbrev = '2_v_5'

for col, dataset in enumerate(datasets):
        unit_colors = {'both': cmaps[net_name][method]['unit_both']}
        mn_colors = {'both': cmaps[net_name][method]['mn_both']}

        df = df_2_v_5_dict[dataset]

        searchnets.plot.metric_v_set_size_df(df=df, 
                                             net_name=net_name,
                                             method=method, 
                                             stimulus=stim_abbrev,
                                             metric='accuracy', 
                                             conditions=['both'],
                                             unit_colors=unit_colors,
                                             mn_colors=mn_colors,
                                             set_ylim=False,
                                             ax=ax[0, col],
                                             ylim=(0.5, 1.05),
                                             add_legend=add_legend)
        ax[0, col].set_title(stim_abbrev + ',\n' + dataset)

        searchnets.plot.metric_v_set_size_df(df=df, 
                                             net_name=net_name,
                                             method=method, 
                                             stimulus=stim_abbrev,
                                             metric='accuracy', 
                                             conditions=['present', 'absent'],
                                             set_ylim=False,
                                             ax=ax[1, col],
                                             ylim=(0.3, 1.05),
                                             add_legend=add_legend)

handles, labels = ax[1, 0].get_legend_handles_labels()
plt.legend(handles, labels, bbox_to_anchor=(0., -0.2, 0.85, .102), loc='upper left',
           ncol=2, mode="expand", borderaxespad=0., fontsize=16);

#savepath = ROOT.joinpath('docs/sfn19/searchstims_alexnet_transfer_initialize_120000samples.png')            
#plt.savefig(savepath, bbox_inches="tight")
```

There are two ways in which the composition of the training set might bias accuracy as measured on the test set. The first and most intuitive to understand is that the stimuli only ever contain one target but can contain up to eight distractors, and so the network will see many more distractors during training, even when the training set is balanced so that there are equal numbers of samples for each visual search set size. One way to deal with this bias would be to only train the network with images of set size 1 so that it sees as many targets as it does distractors. The second way the training set might bias accuracy is that for larger visual search set sizes there is a combinatorial explosion of possible placements of distractors--the network might "memorize" how to respond for distractors in some positions but see many less examples of distractors in other positions. One way to deal with this bias would be to increase the number of examples for larger set sizes.
The above graph tests whether either of these attempts to deal with bias affects accuracy. Specifically we asked how the composition of the training set affects accuracy measured on the test set for the 2 v. 5 visual search stimulus. We chose this stimulus becuase for all models it produced the greatest impairment in accuracy as the number of distractors increased. In the left panel, the training set contained an increasing number of samples for each visual search set size, i.e. [800, 1600, 3200, 6400] samples respectively for set sizes [1, 2, 4, 8]. In the middle panel, the training set contained 3200 samples for each set size, i.e. the dataset was balanced across set sizes (and target present / absent condition). Lastly in the right panel, the training set consisted only of stimuli with visual search set size 1, so that the network saw the exact same number of targets and distractors.
We noted the following: for the left panel, accuracy was always higher for the target absent condition. In comparison, accuracies were roughly the same when samples were balanced across set sizes, as shown in the middle panel. These results taken together suggest that increasing the number of samples for larger set sizes actually impairs accuracy on the target present condition (results shown are for AlexNet but were similar with other models). From the right panel it is very clear that training with just set size of 1 dramatically impaired accuracy in the target present condition. Hence we chose to balance across visual search set sizes for all training.
