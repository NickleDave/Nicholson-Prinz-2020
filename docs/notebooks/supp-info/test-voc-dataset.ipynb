```python
import matplotlib.pyplot as plt
import numpy as np

from searchnets.datasets import VOCDetection
from searchnets.utils.transforms import VOCTransform
```

```python
root = '/home/bart/Documents/data/voc/'
csv_file = '../../../data/Visual_Search_Difficulty_v1.0/VSD_dataset_split.csv'
```

```python
trainset = VOCDetection(root=root,
                        csv_file=csv_file,
                        image_set='trainval',
                        split='train',
                        download=True,
                        transforms=VOCTransform())
```

```python
itertrain = iter(trainset)
```

```python
img, target = next(itertrain)
img = img.numpy()
img = np.transpose(img, (1,2,0))

fig, ax = plt.subplots()
ax.imshow(img)
print(img.shape)
print(target)
```
