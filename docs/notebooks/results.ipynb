# Behavior of convolutional networks performing a frequently used visual search task

## Introduction

### Conectionist models of cognition and deep neural networks
Since the 1980s, artificial neural networks trained with backpropagation gave rise to Connectionist models of cognition. Similar neural network models lie at the heart of the recent rise of deep learning, in large part because of advances in methods for training algorithms, and in technology such as parallelized training on GPUs. When benchmarked with scalar metrics like accuracy on computer vision tasks such as object recognition, these deep neural networks appear capable of human or super-human levels of skill.

### How much can accuracy on a task tell us about cognition?
The impressive accuracy of deep convolutional neural networks raises the question of how these new-and-improved networks perform in other tasks. 

### A case study: the classical form of the visual search task
The "classical" form of the visual search paradigm:
- present subject a stimulus containing discrete items, i.e. objects
- ask subjects to report whether a target is present among distractors; experimenter sets probability that the target will be present on any given trial
- measure reaction time and/or accuracy
- often, plot reaction time as a function of set size (the number of discrete objects in stimulus)
